```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

# Prepare Data

```python
df = pd.read_csv("F://Uni/991/Machine_Learning/ML_HW4/Datasets/2/pima_indians_diabetes.csv")
df = df.sample(frac=1 ,replace=False, axis = 0).reset_index(drop=True)
df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import normalize
```

```python
x = df.drop(columns=['class'])
#x = normalize(x)
y = df['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

# Random Forest

```python
estlist = list([5,10,15,50])
featurelist = list([2,4,6,8])
depthlist = list([5,10,15,20])
maxacc=0
model = np.zeros(3)
for est in estlist:
    for feature in featurelist:
        for depth in depthlist:
            rfclf = RandomForestClassifier(n_estimators=est, max_features=feature, max_depth=depth)
            rfclf.fit(x_train, y_train)
            
            print('for n_estimators= ', est, 'for max_features= ', feature, 'for max_depth= ', depth)
            
            y_pred = rfclf.predict(x_train)
            acc = accuracy_score(y_train, y_pred)
            f1s = f1_score(y_train, y_pred, average=None)
            print('Random Forest Accuracy for Train: ', acc)
            print('Random Forest F1-Measure for Train: ', f1s)
            print('---------------')
            
            
            y_pred = rfclf.predict(x_test)
            acc = accuracy_score(y_test, y_pred)
            f1s = f1_score(y_test, y_pred, average=None)
            if acc>maxacc:
                maxacc=acc
                model=[est,feature,depth]
            print('Random Forest Accuracy for Test: ', acc)
            print('Random Forest F1-Measure for Test: ', f1s)
            print('---------------------------------------------------------------')
print('Best Accuracy for Test: ', maxacc, 'for n_estimators= ', model[0], 'for max_features= ', model[1], 'for max_depth= ', model[2])
```

# Three Models

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import svm
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
```

## SVM

```python
degree = list([5])
coef0list = list([2,2.8,3,3.2,3.5,4,5,6])
#remember to change the numbers..12,12 is just like 3,3???
for d in degree:
    for c in coef0list:
        polynomial_svc = svm.SVC(kernel='poly', degree=d, coef0=c)
        polynomial_svc.fit(x_train, y_train)
        y_pred = polynomial_svc.predict(x_test)
        acc = accuracy_score(y_test, y_pred)
        f1s = f1_score(y_test, y_pred, average=None)

        print('Polynomial SVM Accuracy: ', acc, 'for Degree= ', d, 'for r= ', c)
        print('Polynomial SVM F1-Measure: ', f1s, 'for Degree= ', d, 'for r= ', c)
        print('---------------------------------------------------------------')
```

## Bagging with SVM

```python
clf = BaggingClassifier(base_estimator=svm.SVC(kernel='poly', degree=5, coef0=5), n_estimators=100, max_samples=50)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred)
f1s = f1_score(y_test, y_pred, average=None)
print('Bagging with SVM Accuracy: ', acc)
print('Bagging with SVM F-Measure: ', f1s)
```

## Logistic Regression

```python
clf = LogisticRegression()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred)
f1s = f1_score(y_test, y_pred, average=None)
print('Logistic Regression Accuracy: ', acc)
print('Logistic Regression F-Measure: ', f1s)
```

## Adaboost with Logistic Regression

```python
clf = AdaBoostClassifier(base_estimator=LogisticRegression(),algorithm='SAMME', n_estimators=100, random_state=0)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred)
f1s = f1_score(y_test, y_pred, average=None)
print('Adaboost with Logistic Regression Accuracy: ', acc)
print('Adaboost with Logistic Regression F-Measure: ', f1s)
```

# Decision Stump

```python
clf = AdaBoostClassifier(n_estimators=300)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred)
f1s = f1_score(y_test, y_pred, average=None)
print('Adaboost with Decision Stump Accuracy: ', acc)
print('Adaboost with Decision Stump F-Measure: ', f1s)
```
