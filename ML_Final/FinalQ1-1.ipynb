```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
def ClusterColor(image, k):
    #Reshaping the Image
    imagepix = image.reshape((image.shape[0] * image.shape[1], 3))
    #KMeans Clustering the Pixels to k Clusters
    model = KMeans(n_clusters = k)
    model.fit(imagepix)
    #Changing Pixel Colors to the k Centroids Colors
    imagenew = np.zeros_like(imagepix)
    for i in range(k):
        imagenew[model.labels_==i] = model.cluster_centers_[i]
    imagenew = imagenew.reshape((image.shape[0] , image.shape[1], 3))
    return imagenew
```

```python
klist = [2,3,4,5,6,10,15,20]
```

# Parrots

```python
#Read Parrots image
parrots = cv2.imread('F://Uni/991/Machine_Learning/ML_991_Final/Dataset/1/1-1/parrots.jpg')
parrots = cv2.cvtColor(parrots, cv2.COLOR_BGR2RGB)
plt.imshow(parrots)
```

```python
fig = plt.figure(figsize=(16,12))
i=1
#using the functuin ClusterColor for K-Means and coloring
for k in klist:
    plt.subplot(2,4,i)
    plt.title(str(k)+' Clusters Parrots')
    plt.axis('off')
    plt.imshow(ClusterColor(parrots,k))
    i+=1
plt.show()
```
